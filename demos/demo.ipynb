```python
%load_ext autoreload
%autoreload 2

from xaj import odeint
from jax import numpy as np
```

```python
def rhs(x, y):
    return np.array([y[1], -y[0]])

x0 = 0
y0 = np.array([0,1])
```

```python
ns = odeint(rhs, x0, y0, 1, atol=1e-6, rtol=0)
```

```python
x  = np.linspace(-5, 5, num=101)
yn = ns(x)[:,0]
ya = np.sin(x)
```

```python
from matplotlib import pyplot as plt

fig, axes = plt.subplots(2,1, figsize=(12,8))

axes[0].scatter(ns.xs, ns.ys[:,0], color='C0')
axes[0].plot(x, yn, '.-', color='C1')
axes[0].plot(x, ya, '--', color='C2')
axes[1].scatter(x, abs(yn-ya), color='k')
```
