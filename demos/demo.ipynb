```python
%load_ext autoreload
%autoreload 2

from xaj import dev
from jax import numpy as np
```

```python
def rhs(x, y):
    return np.array([y[1], -y[0]])

x0 = 0
y0 = np.array([0,1])
```

```python
odeint = dev.ODEInt(rhs, x0, y0, 1)
```

```python
x = np.linspace(-5, 5, num=101)
y = odeint(x)[:,0]
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(12,8))

plt.scatter(odeint.xs, odeint.ys[:,0], color='C1')
plt.plot(x, y, '.-')
```
